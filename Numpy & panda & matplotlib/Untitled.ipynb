```python
#compute the x and y co-ordinates for the points on the sine curve
x = np.arange(0, 4*np.pi,0.1)
y = np.sin(x)
plt.title('Sine curve')

#plot the points using matplotlib
plt.plot(x,y,'g')
plt.show()
```

```python
#compute the x and y co-ordinates for the points on the sine and cosine curve
x = np.arange(0, 5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#set up a subplot grid that has height 2 and weidth 1,
# and set the first such plot as active
plt.subplot(2,1,1)

#make the first plot
plt.plot(x,y_sin,'g--')
plt.title("Sine Curve")

#set the secound plot as active and make the secound plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'r--')
plt.title('cosine curve')

#show the figure
plt.show()
```

```python
#bar plot

x = [2,8,10]
y = [11,13,16]

x2 = [3,9,11]
y2 = [6,25,7]

plt.bar(x,y)
plt.bar(x2,y2,color = 'g')
plt.title("Bra Graph")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```
